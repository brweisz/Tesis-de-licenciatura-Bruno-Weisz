```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("last_meditions.csv")
df
df_simple = df[df["program_name"] != "multiple_range_checks_u8"]
df_simple = df_simple[df_simple["program_name"] != "memory_writes"]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='program_name', y='elapsed_time', hue='backend', data=df_simple)
plt.title('Time Measurements by program and backend')
plt.xlabel('program')
plt.ylabel('Time')

plt.legend(title='Backend')
plt.show()
```
